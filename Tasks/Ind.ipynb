## Задание

На бинарном изображении найти контуры, вывести их количество и среднюю площадь контура, а также визуализировать контуры на исходном изображении.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)
ret,thresh = cv2.threshold(img,127, 255, cv2.THRESH_BINARY)
contours, hierarchy = cv2.findContours(thresh, 5, 5)
cnt = contours[0] # создание массива точек контура

# Количества контуров
print("Количество контуров: ", len(contours))

# Расчет площадей контуров
areas = []
for cnt in contours:
    areas.append(cv2.contourArea(cnt))

# Вывод средней площади контура
mean_area = np.mean(areas)
print("Средняя площадь контура: ", mean_area)

# Визуализация контуров
img_contours = np.zeros_like(img)
cv2.drawContours(img_contours, contours, -1, 255, 2)

# Вывод изображений
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.xticks([])
plt.yticks([])

plt.subplot(122)
plt.imshow(img_contours, cmap='gray')
plt.title('Contours')
plt.xticks([])
plt.yticks([])

plt.show()
```
